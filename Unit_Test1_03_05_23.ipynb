```python
gender_counts = df.groupby('Gender')['User_ID'].count()


plt.bar(['Male', 'Female'], gender_counts.values, color=['black', 'pink'])
plt.title('Number of buyers by gender',size=25)
plt.xlabel('Gender')
plt.ylabel('Number of buyers')
plt.show()
```

```python
print(gender_counts) #her we can see the female buyers are more tha male buyers.
```

```python
import matplotlib.pyplot as plt


age_data = df['Age']


plt.hist(age_data, bins=20)

plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
#as we can see 30 aged peoples are the mejority buyers
```

```python
highest_sale=df.groupby('State')['Amount'].sum().reset_index()
highest_sale
```

```python
plt.figure(figsize=(15,5))
sns.set_style('darkgrid')
sns.countplot(x=df['State'],palette='Set2',edgecolor='black')
plt.title("highest sales as per State" ,size=30)
plt.xlabel("States",size=15)
plt.xticks(size=12)

plt.ylabel("Amount",size=15 )
plt.yticks(size=12)
plt.show()
```
